```python
#set up

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

from skimage.color import rgb2gray
```

## Read, display, and write images

```python
grayIm = mpimg.imread('buckeyes_gray.bmp')
plt.imshow(grayIm, cmap = 'gray')
plt.axis('image')
plt.imsave('bucyeyes_gray.jpg', grayIm, cmap = 'gray')
```

```python
rgbIm = mpimg.imread('buckeyes_rgb.bmp')
plt.imshow(rgbIm)
plt.axis('image')
plt.imsave('bucyeyes_rgb.jpg', rgbIm)
```

## Convert rgb image to grayscale

```python
grayImFromRGB = rgb2gray(rgbIm)
plt.imshow(grayImFromRGB, cmap = 'gray')
plt.axis('image')
```

## Creating, writing, and reading a checker-board image

```python
zBlock = np.zeros([10,10])
oBlock = np.ones([10,10])*255
r1 = np.hstack((zBlock, oBlock))
r2 = np.hstack((oBlock, zBlock))
pattern = np.vstack((r1,r2))
checkerIm = np.tile(pattern,(5,5))
plt.imsave('checkerIm.bmp', np.uint8(checkerIm), cmap = 'gray')
```

```python
Im = mpimg.imread('checkerIm.bmp')
plt.imshow(Im)
plt.axis('image')
```
